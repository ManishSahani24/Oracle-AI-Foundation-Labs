```python
from sklearn.datasets import make_circles
import numpy as np
import matplotlib.pyplot as plt  
```

**make_circles** - generates a synthetic 2D dataset of points arranged in two concentric circles. 
                    When noise = 0, datapoints are arranged along the circumference of both the concentric circles.

```python
X, y = make_circles(n_samples = 300, noise = 0.1, factor = 0.5, random_state=0)
```

**factor** - It is the ratio of inner circle radius to the outer circle radius.

**random_state** - for reproducibility

```python
plt.figure(figsize = (8,6))
plt.scatter(X[:, 0], X[:, 1],c=y,cmap = plt.cm.RdYlGn, edgecolors = 'k',marker='o', s=50)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Plot of points")
plt.colorbar(label = 'Class Label')
plt.show()
```
